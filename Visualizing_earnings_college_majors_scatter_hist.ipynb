# Investigating Job Outcomes for College Graduates Between 2010 and 2012 Using Data Visualizations

Herein we will use a variety of data visualizations to answer questions about job outcomes for students who graduated from college between 2010 and 2012.

Though the original dataset was released by American Community Survey, which conducts surveys and aggregates data, FiveThirtyEight cleaned the dataset and released it their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

We will explore questions like:

* Do students in more popular majors make more money?
  * Using scatter plots
* How many majors are predominantly male? Predominantly female?
  * Using histograms
* Which category of majors have the most students?
  * Using bar plots 
  
Each row in the dataset represents a different major in college and contains the following columns we will be looking at:

| Column Name    | Description                                                     |
|----------------|-----------------------------------------------------------------|
| Rank           | Rank by median earnings (the dataset is ordered by this column) |
| Major_code     | Major code                                                      |
| Major          | Major description                                               |
| Major_category | Category of major                                               |
| Total          | Total number of people with major                               |
| Sample_size    | Sample size (unweighted) of full-time                           |
| Men            | Male graduates                                                  |
| Women          | Female graduates                                                |
| ShareWomen     | Women as share of total                                         |
| Employed       | Number employed                                                 |
| Median         | Median salary of full-time, year-round workers                  |
| Low_wage_jobs  | Number in low-wage service jobs                                 |
| Full_time      | Number employed 35 hours or more                                |
| Part_time      | Number employed less than 35 hours                              |

```python
# imports pandas and matplot
import pandas as pd
import matplotlib.pyplot as plt

# Read in the graduation csv file into pandas DataFrame
recent_grads = pd.read_csv('recent-grads.csv')

# Jupyter magic code to display plots inline
%matplotlib inline
```

## Exploring relationships with scatter plots

```python
# Generates and displays a Sample_size vs. Median scatter plot
recent_grads.plot(x='Sample_size', y='Median', kind='scatter',\
                  title='Sample Size vs. Median salary', figsize=(10,10))
```

It looks like as majors with larger sample sizes have a lower median salaries than those with smaller sample sizes.

```python
# Generates and displays a Sample_size vs. Unemployment_rate scatter plot
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter',\
                  title='Sample Size vs. Unemployment Rate', figsize=(10,10))
```

Majors wth larger sample sizes appear to converge on a 5% to 7% unemployment rate

### Is there any link between the number of full-time employees and median salary?

```python
# Generates and displays a Full_time vs. Median scatter plot
recent_grads.plot(x='Full_time', y='Median', kind='scatter',\
                  title='number of full time employees vs. Median salary',\
                  figsize=(10,10))
```

It would seem that majors with less full time employees make more money than majors with hundreds of thousands of employees.

```python
# Generates and displays a ShareWomen vs. Unemployment_rate scatter plot
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter',\
                  title='Women as share of total vs Unemployment rate',\
                  figsize=(10,10))
```

There is a high variance between unemployment rate and the percentage of employed women as a share of the total employment numbers. There is no discernible correlation between these two variables.

```python
# Generates and displays a Men vs. Median scatter plot
recent_grads.plot(x='Men', y='Median', kind='scatter',\
                  title='Male graduates vs Median salary', figsize=(10,10))
```

It would seem that the majority of male graduates make a median salary between \$20,000 and \$60,000 a year, with a great deal making around \$40,000 per year.

```python
# Generates and displays a Women vs. Median scatter plot
recent_grads.plot(x='Women', y='Median', kind='scatter',\
                  title='Female graduates vs Median Salary', figsize=(10,10))
```

It would seem that the majority of female graduates make between \$20,000 and \$60,000 a year, with a great deal making around \$30,000 to \$40,000 per year.

### Do students in more popular majors make more money?

```python
# Generates and displays a Total vs. Median scatter plot
recent_grads.plot(x='Total', y='Median', kind='scatter',\
                  title='Total number of students in major vs Median Salary',\
                  figsize=(10,10))
```

It would seem that the most popular majors do not make the most money, and that actually students with the most popular majors make around $40,000 per year or less. 

Majors with less students might be popular as well due to their higher earning potential but could be more selective with their acceptance criteria, given the academic rigor required to complete them (i.e. engineering)

### Do students that majored in subjects that were majority female make more money?

```python
# Generates and displays a ShareWomen vs. Median scatter plot
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter',\
                  title='Women as share of total vs Median salary',\
                  figsize=(10,10))
```

It looks like students that majored in subjects that were majority female (greater than 0.5) do not make more money. According to this dataset, they make a median salary of between \$20,000 and \$50,000 per year. Meanwhile, majors that are majority male make a median salary between \$30,000 and \$75,000 a year.

There is a weak negative correlation between medain salary and the percent of woman in the major, majors with a higher percentage of woman see a lower medan salary.

## Using bar plots to compare majors

```python
# Displays a plot to compare the percentage of women from the first 10 rows
recent_grads.head(10).plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
# Displays a plot to compare the percentage of women from the last 10 rows
recent_grads.tail(10).plot.bar(x='Major', y='ShareWomen', legend=False)
```

Given that the dataset is ranked by median salary in descending order, it appears that women make up the majority of less lucrative majors.

```python
# Displays a plot to compare the unemployment rate of the first 10 rows
recent_grads.head(10).plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

```python
# Displays a plot to compare the unemployment rate of the last 10 rows
recent_grads.tail(10).plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

It appears that the unemployment rates indiscriminate of median salary earnings, since we find high unemployment rates of both ends of the median salary spectrum. 

Perhaps unemployment rates are higher for majors belonging to niche fields such as nuclear engineering and clinical psychology.

## Grouped bar plot, box plot and Hexagonal bin plot to investigate a few more relationships

*  Grouped bar plot to compare the number of men with the number of women in each category of majors
* Box plot to explore the distributions of median salaries and unemployment rate
* Hexagonal bin plot to visualize the columns that had dense scatter plots


```python
# Adds number of men and woman in each major category and displays them as 
# a grouped bar plot
recent_grads.groupby("Major_category")["Women","Men"].sum().plot.bar(figsize=(10,5))
```

It looks like Business is the most popular category of majors for both men and women, while interdisciplinary studies are the least popular major category.

We could use more men in education, communication & journalism, and health and psychology & social work and more women in engineering and computers & mathematics to have a more balanced gender diversity within universities.

```python
# Displays a box pot for median salaries
recent_grads[["Median"]].boxplot(figsize=(10,5))
```

The median median salary across all majors hovers just abouve \$35,000 a year. While most median salaries range from ~\$22,000 to \$62,000 per year.

```python
# Displays a box pot unemployment rates
recent_grads[["Unemployment_rate"]].boxplot(figsize=(10,5))
```

The median unemployment rate across all majors is around 7%, with most majors ranging from 0% to 13%.

```python
# Displays a hexagonal bin plot to better visualize
# total students vs median income
recent_grads.plot.hexbin(x='Total', y='Median', gridsize=30, figsize=(10,5),
                        cmap='plasma',sharex=False)
```

We can confirm that the majority of majors have less than 50,000 students and earn a median slary of \$35,000 to \$40,000 a year.

In this project, We explored data on earnings and job outcomes for college majors using a variety of plotting/visualization tools built into pandas.

We found that:
